```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.jitimu.com/2023/02/nhif-hospitals-offering-renal-dialysis"

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
results = soup.find("table", style="width: 100%;")
print(results)
```

```python
soup.find_all("table")
```

```python
headers = results.find_all("th")
print(headers)
```

```python
table= soup.find_all('table')

print(table)
```

```python
diabetes_facilities= [title.text for title in headers]

print(diabetes_facilities)
```

```python
# Getting the dataframe
import pandas as pd
df = pd.DataFrame(columns = diabetes_facilities)
df
```

```python
column_data = results.find_all('tr')
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]


    length = len(df)
    df.loc[length] = individual_row_data
   
```
